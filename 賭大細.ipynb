```python
#Step 1 - Buy in

def updatePrincipal(principal):
    money = input("Please input the money you want to add:")
    principal = principal + int(money)
    print ("You have bought-in $" + str(money))
    return principal
```

```python
#Step 5 - Rolling 3 dice

import random

def rollDice():
    rollResults = []

    for num in range(3):
        roll = random.randint(1, 6)
        rollResults.append(roll)

    for diceNum in rollResults:
        getDiceTUI(diceNum)
        
    return rollResults
```

```python
#Step 6 - Compare roll results

def compareResults(principal, bet, betBigSmall, rollResults):
    sum_of_roll_results = sum(rollResults)
    print ("Sum of dice: " + str(sum_of_roll_results))
    checkTriple = all(element == rollResults[0] for element in rollResults)
    if checkTriple == True:
        print("Triple. Dealer wins")
        principal = principal - bet
    elif sum_of_roll_results >= 11:
        print ("Opened big")
        if betBigSmall == "B":
            principal = principal + bet
        else:
            principal = principal - bet
    else:
        print ("Opened small")
        if betBigSmall == "S":
            principal = principal + bet
        else:
            principal = principal - bet
    return principal
```
